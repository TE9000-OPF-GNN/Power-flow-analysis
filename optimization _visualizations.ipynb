```python

import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
#visualizations of optimizations


# Define the function z = f(x, y)
def func(x, y):
    return np.sin(np.sqrt(x**2 + y**2)) + 0.1 * (x**2 + y**2)

# Create a grid of x and y values
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)

# Compute z values
z = func(x, y)

# Create the 3D plot
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')

# Plot the surface
surf = ax.plot_surface(x, y, z, cmap='viridis', edgecolor='none', alpha=0.8)

# Add a color bar for reference
fig.colorbar(surf, ax=ax, shrink=0.5, aspect=10)

# Add labels
ax.set_title("3D Surface Plot with Local and Global Minima")
ax.set_xlabel("X-axis")
ax.set_ylabel("Y-axis")
ax.set_zlabel("Z-axis")

# Show the plot
plt.show()
```

```python
def plot_surface(f, title, cmap):
    x = np.linspace(-2, 2, 100)
    y = np.linspace(-2, 2, 100)
    x, y = np.meshgrid(x, y)
    z = f(x, y)

    fig = plt.figure(figsize=(10, 7))
    ax = fig.add_subplot(111, projection='3d')
    surf = ax.plot_surface(x, y, z, cmap=cmap, edgecolor='none')
    ax.set_title(title)
    ax.set_xlabel('X-axis')
    ax.set_ylabel('Y-axis')
    ax.set_zlabel('Z-axis')
    fig.colorbar(surf, shrink=0.5, aspect=10)
    plt.show()

# Positive definite Hessian (local minimum)
def f_min(x, y):
    return x**2 + y**2

plot_surface(f_min, 'f(x,y)= x^2+y^2: Positive Definite Hessian: Local Minimum', 'viridis')

# Negative definite Hessian (local maximum)
def f_max(x, y):
    return -x**2 - y**2

plot_surface(f_max, 'g(x,y)=-x^2-y^2: Negative Definite Hessian: Local Maximum', 'plasma')

# Indeterminate Hessian (saddle point)
def f_saddle(x, y):
    return x**2 - y**2

plot_surface(f_saddle, 'h(x,y)=x^2-y^2: Indeterminate Hessian: Saddle Point', 'coolwarm')
```
